```python
import matplotlib.pyplot as plt

# بيانات اللاعبين
players_positions = {
    'Goalkeeper': [5, 30],
    'Right Back': [25, 45],
    'Left Back': [25, 15],
    'Center Back': [25, 30],
    'Midfielder': [50, 30],
    'Right Winger': [70, 45],
    'Left Winger': [70, 15],
    'Striker': [85, 30]
}

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('green')
ax.set_xlim(0, 100)
ax.set_ylim(0, 60)
ax.axis('off')

# خطوط الملعب
ax.plot([0, 100], [0, 0], color='white')
ax.plot([0, 100], [60, 60], color='white')
ax.plot([0, 0], [0, 60], color='white')
ax.plot([100, 100], [0, 60], color='white')
ax.plot([50, 50], [0, 60], color='white', ls='--')

# حفظ العناصر المرسومة
texts = {}
for name, (x, y) in players_positions.items():
    texts[name] = ax.text(x, y, name, ha='center', va='center', fontsize=10,
                          bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3', alpha=0.7))

# تحريك العناصر بالماوس
dragging = {'text': None}

def on_press(event):
    for name, text in texts.items():
        contains, _ = text.contains(event)
        if contains:
            dragging['text'] = text
            break

def on_motion(event):
    if dragging['text'] is not None and event.xdata and event.ydata:
        dragging['text'].set_position((event.xdata, event.ydata))
        fig.canvas.draw()

def on_release(event):
    dragging['text'] = None

fig.canvas.mpl_connect('button_press_event', on_press)
fig.canvas.mpl_connect('motion_notify_event', on_motion)
fig.canvas.mpl_connect('button_release_event', on_release)

plt.title("Drag players to move them", fontsize=14, color='white')
plt.show()
```

```python
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, VBox

# أسماء اللاعبين ومواقعهم الافتراضية
players = {
    'gk': [5, 30],
    'Advit': [25, 45],
    'Arjan': [25, 15],
    'Pito': [25, 30],
    'Adish': [50, 30],
    'Aryan': [70, 45],
    'Adil': [70, 15],
    
}

# رسم الملعب مع أماكن اللاعبين
def draw_board(**kwargs):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_facecolor('green')
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 60)
    ax.axis('off')

    # خطوط الملعب
    ax.plot([0, 100], [0, 0], color='green')
    ax.plot([0, 100], [60, 60], color='green')
    ax.plot([0, 0], [0, 60], color='green')
    ax.plot([100, 100], [0, 60], color='green')
    ax.plot([50, 50], [0, 60], color='gree', ls='--')

    # رسم أسماء اللاعبين
    for name in players:
        x = kwargs[f'{name}_x']
        y = kwargs[f'{name}_y']
        ax.text(x, y, name, ha='center', va='center', fontsize=10,
                bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3', alpha=0.7))

    plt.title("Tactical Board - Move Players", fontsize=14, color='white')
    plt.show()

# إنشاء السلايدرات لجميع اللاعبين
sliders = {}
for name, (x, y) in players.items():
    sliders[f'{name}_x'] = FloatSlider(value=x, min=0, max=100, step=1, description=f'{name} X')
    sliders[f'{name}_y'] = FloatSlider(value=y, min=0, max=60, step=1, description=f'{name} Y')

# تشغيل الواجهة التفاعلية
interact(draw_board, **sliders)
```
